```python
import numpy as np
import scipy
import matplotlib.pyplot as plt
import scipy.stats as stats
import statsmodels
import statsmodels.sandbox.stats.multicomp
import pandas as pd
from statsmodels.stats.weightstats import ztest

import seaborn as sns
```

> ### 1.а. Проверить работу критериев MW, Z, T на:
- выборках с выбросами размера 1000

```python
# Генерация выборок с 1000 случайных чисел из нормального распределения

sample_A = np.random.normal(loc=0, scale=1, size=1000)
sample_B = np.random.normal(loc=0, scale=1, size=1000)
```

```python
# Добавление выбросов

outliers = np.random.uniform(low=-10, high=10, size=10)
sample_A = np.concatenate((sample_A, outliers))
sample_B = np.concatenate((sample_B, outliers))
```

```python
# Создание гистограмм двух выборок sample_A и sample_B и их визуализации на графике

plt.figure(figsize=(7,7))
plt.hist(sample_A, bins = 32, color='green',  linewidth=2)
plt.hist(sample_B, bins = 32, color='red',  linewidth=2)
```

```python
# Смотрим среднее

print(sample_A.mean(), sample_B.mean())
```

```python
# Смотрим стандартное отклонение

print(sample_A.std(), sample_B.std())
```

Сравниваем результаты, полученные на разных стат. критериях

```python
# Проверка критерия Манна-Уитни
mannwhitney_u, mannwhitney_p = stats.mannwhitneyu(sample_A, sample_B)
print(f"Mann-Whitney U: {mannwhitney_u}")
print(f"Mann-Whitney p-value: {mannwhitney_p}" '\n')

# Проверка Z-теста
z_stat, z_p = ztest(sample_A, sample_B)
print(f"Z-statistic: {z_stat}")
print(f"Z p-value: {z_p}" '\n')

# Проверка T-теста
t_stat, t_p = stats.ttest_ind(sample_A, sample_B)
print(f"T-statistic: {t_stat}")
print(f"T p-value: {t_p}")
```

> ### 2. Проверить выполнение ЦПТ для распределений (надо написать функцию):
- нормального
- биномиального
- экспоненциального
- гамма

Проверять каждое для случая 100 и 1000 наблюдений в выборке.

При необходимости можно пользоваться критериями для проверки нормальности распределения

Сделать выводы

```python
import numpy as np
from scipy.stats import shapiro, ttest_ind, kstest, gamma, binom, norm, expon
import matplotlib.pyplot as plt
```

```python
import numpy as np
import matplotlib.pyplot as plt

def check_cpt_distributions(distributions, sample_size):
    for distribution in distributions:
        if distribution == "normal":
            # Генерация выборки из нормального распределения
            sample = np.random.normal(loc=0, scale=1, size=sample_size)
        elif distribution == "binomial":
            # Генерация выборки из биномиального распределения
            sample = np.random.binomial(n=10, p=0.5, size=sample_size)
        elif distribution == "gamma":
            # Генерация выборки из гамма распределения
            sample = np.random.gamma(shape=2, scale=1, size=sample_size)
        elif distribution == "exponential":
            # Генерация выборки из экспоненциального распределения
            sample = np.random.exponential(scale=1, size=sample_size)
        else:
            print(f"Unknown distribution: {distribution}")
            continue

        # A/B-тестирование
        _, p_value = ttest_ind(sample, np.zeros(sample_size))

        # Проверка нормальности распределения выборки
        _, p_value_normality = shapiro(sample)

        # Проверка соответствия распределения выборки теоретическому распределению
        if distribution == "normal":
            p_value_theoretical = kstest(sample, norm(loc=0, scale=1).cdf)[1]
        elif distribution == "binomial":
            p_value_theoretical = kstest(sample, binom(n=10, p=0.5).cdf)[1]
        elif distribution == "gamma":
            p_value_theoretical = kstest(sample, gamma(a=2, scale=1).cdf)[1]
        elif distribution == "exponential":
            p_value_theoretical = kstest(sample, expon(scale=1).cdf)[1]

        # Вывод результатов
        print(f"Distribution: {distribution}")
        print(f"Sample size: {sample_size}")
        print(f"A/B Test p-value: {p_value}")
        print(f"Shapiro-Wilk test p-value: {p_value_normality}")
        print(f"Kolmogorov-Smirnov test p-value (theoretical): {p_value_theoretical}")
        print("")

        # Визуализация распределения
        plt.figure()
        plt.hist(sample, bins=30, density=True, alpha=0.6, label="Sample")
        plt.xlabel("Value")
        plt.ylabel("Density")
        plt.title(f"Distribution: {distribution} (Sample size: {sample_size})")
        plt.legend()
        plt.show()

# Проверка ЦПТ для выборок размером 100 и 1000, для нормального, биномиального, гамма и экспоненциального распределений
distributions = ["normal", "binomial", "gamma", "exponential"]
sample_size = 100
check_cpt_distributions(distributions, sample_size)

sample_size = 1000
check_cpt_distributions(distributions, sample_size)
```

## Выводы:

### При выборке в 100 наблюдений:

1. Для нормального распределения ЦПТ выполняется - распределение средних не отличается значимо от нормального согласно критериям.
2. Для биномиального и гамма распределений ЦПТ не выполняется - есть значимые отличия от нормального распределения.
3. Для экспоненциального распределения результат неоднозначен - по К-С тесту оно близко к нормальному, а по Шапиро-Уилка есть отличия.

### При выборке в 1000 наблюдений:

4. Для нормального распределения ЦПТ выполняется - распределение средних не отличается от нормального.

- -- Генерируем выборку размером 200 из гамма-распределения с параметрами shape=5, scale=1.
- -- Так как данные непрерывные и не нормально распределены, выбираем непараметрический критерий - Манна-Уитни.
- -- Разделим выборку на 2 части по 100 элементов в каждой (контрольная и тестовая).
- -- Просчитаем U статистику Манна-Уитни для этих выборок.
- -- Оценим мощность критерия при заданном эффекте (например, сдвиг среднего на 10%) с помощью моделирования.
- -- Проверим долю ошибок 1-го рода, генерируя выборки из одного распределения.
- -- Построим гистограмму распределения p-value, оно должно быть равномерным.

```python
# Генерируем выборку из гамма-распределения
data = np.random.gamma(5, 1, 200)

# Разбиваем на 2 выборки
control = data[:100]
test = data[100:]

# Применяем критерий Манна-Уитни
u, p = mannwhitneyu(control, test)

# Оценка мощности при сдвиге среднего на 10%
mean_diff = 0.1 * np.mean(control)
power = []
for i in range(1000):
  control = np.random.gamma(5, 1, 100)
  test = np.random.gamma(5, 1, 100) + mean_diff
  u, p = mannwhitneyu(control, test)
  power.append(p < 0.05)
print("Мощность:", np.mean(power))

# Проверка доли ошибок 1-го рода
errors = []
for i in range(1000):
  sample = np.random.gamma(5, 1, 200)
  control = sample[:100]
  test = sample[100:]
  u, p = mannwhitneyu(control, test)
  errors.append(p < 0.05)
print("Доля ошибок:", np.mean(errors))

# Гистограмма распределения p-value
plt.hist(p, bins=20)
plt.show()
```

- -- Мощность критерия получилась невысокой - всего 35.1%. Это значит, что при сдвиге среднего на 10% критерий Манна-Уитни сможет его обнаружить только в 35% случаев.
- -- Низкая мощность связана с небольшим размером выборки (по 100 наблюдений). - -- Для повышения мощности нужно увеличить объем данных.

- -- Таким образом, критерий Манна-Уитни при данном объеме данных работает корректно, но имеет невысокую мощность.
